```python
%matplotlib inline

import matplotlib.pyplot as plt
from IPython.display import clear_output
import time


# USER INPUT

user_input = input(
    "Enter numbers separated by commas: "
)

split_numbers = user_input.split(",")

numbers = [

    int(num)

    for num in split_numbers

]


# SPEED INPUT

speed = float(

    input(
        "Enter animation speed (example 0.5): "
    )

)


# BUBBLE SORT FUNCTION

def bubble_sort(numbers):

    steps = []

    steps.append(numbers.copy())

    for i in range(len(numbers)):

        for j in range(len(numbers) - 1):

            if numbers[j] > numbers[j + 1]:

                temp = numbers[j]

                numbers[j] = numbers[j + 1]

                numbers[j + 1] = temp

                steps.append(numbers.copy())

    return steps


# SELECTION SORT FUNCTION

def selection_sort(numbers):

    steps = []

    steps.append(numbers.copy())

    for i in range(len(numbers)):

        min_index = i

        for j in range(i + 1, len(numbers)):

            if numbers[j] < numbers[min_index]:

                min_index = j

        temp = numbers[i]

        numbers[i] = numbers[min_index]

        numbers[min_index] = temp

        steps.append(numbers.copy())

    return steps


# CREATE COPIES

bubble_numbers = numbers.copy()

selection_numbers = numbers.copy()


# TIME BUBBLE SORT

start = time.time()

bubble_steps = bubble_sort(
    bubble_numbers
)

bubble_time = time.time() - start


# TIME SELECTION SORT

start = time.time()

selection_steps = selection_sort(
    selection_numbers
)

selection_time = time.time() - start


# PRINT PERFORMANCE

print(
    "Bubble Sort Time:",
    bubble_time
)

print(
    "Selection Sort Time:",
    selection_time
)


# VISUALIZATION

max_steps = max(
    len(bubble_steps),
    len(selection_steps)
)

for i in range(max_steps):

    clear_output(wait=True)

    fig, axes = plt.subplots(
        1,
        2,
        figsize=(10,4)
    )


    # BUBBLE SORT GRAPH

    if i < len(bubble_steps):

        axes[0].bar(

            range(len(bubble_steps[i])),

            bubble_steps[i],

            color="blue"

        )

        axes[0].set_title(
            "Bubble Sort"
        )


    # SELECTION SORT GRAPH

    if i < len(selection_steps):

        axes[1].bar(

            range(len(selection_steps[i])),

            selection_steps[i],

            color="green"

        )

        axes[1].set_title(
            "Selection Sort"
        )


    plt.show()

    time.sleep(speed)
```
