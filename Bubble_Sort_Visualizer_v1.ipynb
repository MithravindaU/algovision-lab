```python
import matplotlib.pyplot as plt
import time
```

```python
numbers = [5, 2, 8, 1]

steps = []

steps.append(numbers.copy())

for i in range(len(numbers)):

    for j in range(len(numbers) - 1):

        if numbers[j] > numbers[j + 1]:

            temp = numbers[j]

            numbers[j] = numbers[j + 1]

            numbers[j + 1] = temp
            steps.append(numbers.copy())

print(numbers)
```

```python
for step in steps:

    print(step)
```

```python
for step in steps:

    plt.clf()

    plt.bar(range(len(step)), step)

    plt.pause(0.5)
plt.show()
```
